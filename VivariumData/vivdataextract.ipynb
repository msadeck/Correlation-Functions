```python
import pickle
import pandas as pd

# Load the pickle file
with open('data_export.pkl', 'rb') as f:
    data = pickle.load(f)

# Extract data into a list of dictionaries
rows = []

for timepoint, timepoint_data in data.items():
    agents = timepoint_data.get('agents', {})
    
    for agent_id, agent_data in agents.items():
        # Extract cell state
        cell_state = agent_data.get('internal', {}).get('cell_state', None)
        
        # Extract coordinates
        location = agent_data.get('boundary', {}).get('location', [None, None])
        x_coord = location[0] if len(location) > 0 else None
        y_coord = location[1] if len(location) > 1 else None
        
        # Determine cell type from agent_id prefix
        cell_type = agent_id.rsplit('_', 1)[0]
        
        rows.append({
            'timepoint': timepoint,
            'cell_id': agent_id,
            'cell_type': cell_type,
            'cell_state': cell_state,
            'x_coord': x_coord,
            'y_coord': y_coord
        })

# Create DataFrame
df = pd.DataFrame(rows)
df = df.sort_values(['timepoint', 'cell_id']).reset_index(drop=True)

# Save to CSV
df.to_csv('extracted_cell_data.csv', index=False)
```
